```python
# Import the required libraries
mport numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from sklearn.datasets import load_breast_cancer
```

```python
# Load the breast cancer dataset imported from sklearn 
data = load_breast_cancer()
X = data.data
y = data.target
print(X.shape, y.shape)
```

```python
# Split the dataset into train and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# Create a SLP model
model_1 = Sequential()
model_1.add(Dense(units=1, input_dim= X.shape[1], activation='sigmoid'))
model_1.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])
history = model_1.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))
loss, accuracy = model_1.evaluate(X_train, y_train)
print(f"Loss: {loss}, Accuracy: {accuracy}")
```

```python
# Plot the loss and accuracy curve for training and validation
import matplotlib.pyplot as plt
train_loss = history.history['loss']
val_loss = history.history['val_loss']
train_acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(train_loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.subplot(1, 2, 2)
plt.plot(train_acc, label='Training Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Training and Validation Accuracy')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD

# Define a range of learning rates to test
learning_rates = np.linspace(0.0001, 0.1, 50)  # 50 values between 0.0001 and 0.1
losses = []  
accuracies = [] 

# Loop over each learning rate to train the model
for lr in learning_rates:
    # Define a MLP model with 16 units in a single hidden layer
    model = Sequential()
    model.add(Dense(units=16, input_dim=X.shape[1], activation='sigmoid'))  # Hidden layer with 16 neurons
    model.add(Dense(units=1, activation='sigmoid'))  # Output layer with 1 neuron for binary classification
    
    # Compile the model with the current learning rate using SGD optimizer
    optimizer = SGD(learning_rate=lr)
    model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
    
    # Train the model on the training data
    history = model.fit(X_train, y_train, epochs=30, verbose=0)  # Train for 30 epochs, no output shown
    
    # Evaluate the model on the test data
    loss, accuracy = model.evaluate(X_test, y_test)
    
    # Store the loss and accuracy for the current learning rate
    losses.append(loss)
    accuracies.append(accuracy)

# Identify the best learning rate 
best_lr_index = np.argmax(accuracies) 
best_lr = learning_rates[best_lr_index]  
best_accuracy = accuracies[best_lr_index] 

# Print the best learning rate and corresponding accuracy
print("Best learning rate:", best_lr)
print("Corresponding accuracy:", best_accuracy)

# Plot the accuracy and loss for different learning rates
plt.figure(figsize=(14, 7))

# Plot the accuracy vs learning rate
plt.subplot(1, 2, 1)
plt.plot(learning_rates, accuracies, label='Accuracy')
plt.scatter(best_lr, best_accuracy, color='red', label='Best Accuracy')  
plt.title('Accuracy vs Learning Rate (SGD)')
plt.xlabel('Learning Rate')
plt.ylabel('Accuracy')
plt.legend()

# Plot the loss vs learning rate
plt.subplot(1, 2, 2)
plt.plot(learning_rates, losses, label='Loss')
plt.title('Loss vs Learning Rate (SGD)')
plt.xlabel('Learning Rate')
plt.ylabel('Loss')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam

learning_rates = np.linspace(0.0001, 0.1, 50)
losses = []
accuracies = []

for lr in learning_rates:
    model = Sequential()
    model.add(Dense(units= 16, input_dim= X.shape[1], activation='sigmoid'))
    model.add(Dense(units=1, activation='sigmoid'))
    optimizer = Adam(learning_rate=lr)
    model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
    history = model.fit(X_train, y_train, epochs=30, verbose=0)
    loss, accuracy = model.evaluate(X_test, y_test)
    losses.append(loss)
    accuracies.append(accuracy)
    print(accuracies)

plt.figure(figsize=(14, 7))
plt.subplot(1, 2, 1)
plt.plot(learning_rates, accuracies)
plt.title('Accuracy vs Learning Rate (Adam)')
plt.xlabel('Learning Rate')
plt.ylabel('Accuracy')
plt.subplot(1, 2, 2)
plt.plot(learning_rates, losses)
plt.title('Loss vs Learning Rate (Adam)')
plt.xlabel('Learning Rate')
plt.ylabel('Loss')
plt.tight_layout()
plt.show()
```
